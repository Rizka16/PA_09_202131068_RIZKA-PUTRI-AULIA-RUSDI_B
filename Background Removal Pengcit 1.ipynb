```python
input_path = 'uas.jpg'
img = Image.open(input_path)
img_name = 'Real_Pict'
```

## Proses Remove Background

```python
output_path = "uas-removed-background.png"
```

```python
with open(output_path, 'wb') as f:
    input_pict = open(input_path,'rb').read()
    subject = remove(input_pict) 
    f.write(subject)
```
